```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = 4 * x - 4

plt.plot(x,y, label='Y = 4X - 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph of Y = 4X -4')
plt.legend()
plt.grid(True)
plt.show()
```
